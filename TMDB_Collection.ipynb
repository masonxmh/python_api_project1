```python
# # Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import sys
# Import API key
from api_keys import api_key
from api_keys2 import api_key2
```

## read 10,000 Movie ID from 2009 ~ 2019 through TMDB API URL

```python
record_counter = 0
movie_id_list = []
base_url= "https://api.themoviedb.org/3/discover/movie?"

print("Beginning Data Retrieval\n")
print("-"*30)

for page in range(500):   
    movie_url=base_url+"primary_release_date.gte=2009-01-01&primary_release_date.lte=2019-10-23&api_key="+api_key+"&page="+str(page+1)
    response=requests.get(movie_url).json()
    for index in range(len(response['results'])):
        movie_id_list.append(response['results'][index]["id"])
        record_counter+=1
        sys.stdout.write('\r'+'Processing Record '+str(record_counter))
        sys.stdout.flush()

print ("\nData Retrieval Complete\n")
print ("-"*30)
print ("Movie Found "+str(record_counter))
print(f"There are {len(movie_id_list)} in the List")
```

## scrap 10,000 records year 2009 to 2019 from TMDB API by ID

```python
#initial counter
record_counter=0
set_counter=1

#define list
id_list=[]
title_list=[]
adult_list=[]
budget_list=[]
genres_list=[]
language_list=[]
pop_list=[]
company_list=[]
country_list=[]
release_date_list=[]
revenue_list=[]
run_time_list=[]
vote_ave_list=[]
vote_count_list=[]
imdb_id_list=[]

base_url = "https://api.themoviedb.org/3/movie/"

print("Beginning Data Retrieval\n")
print("-"*30)
#Scrap data from api
for m_id in movie_id_list:
    try:
        url=base_url+str(m_id)+"?api_key="+api_key#+"&language=en-US"
        response=requests.get(url).json()
        id_list.append(response["id"])
        title_list.append(response["title"])
        adult_list.append(response["adult"])
        budget_list.append(response["budget"])
        genres_list.append([i["name"] for i in response["genres"]])
        language_list.append(response["original_language"])
        pop_list.append(response["popularity"])
        company_list.append([j["name"] for j in response["production_companies"]])
        country_list.append([k["name"] for k in response["production_countries"]])
        release_date_list.append(response["release_date"])
        revenue_list.append(response["revenue"])
        run_time_list.append(response["runtime"])
        vote_ave_list.append(response["vote_average"])
        vote_count_list.append(response["vote_count"])
        imdb_id_list.append(response["imdb_id"])
        
        record_counter+=1
        sys.stdout.write('\r'+'Processing Record '+str(record_counter))
        sys.stdout.flush()

    except:
        pass

print ("\nData Retrieval Complete\n")
print ("-"*30)
print ("Movie Found "+str(record_counter))
```

### DataFram of tmdb_movie_df

```python
tmdb_movie_df=pd.DataFrame({"ID":id_list,
                       "Title": title_list,
                       "Adult":adult_list,
                       "Budget":budget_list,
                       "Genres":genres_list,
                       "Language":language_list,
                       "Production Company":company_list,
                       "Popularity":pop_list,
                       "Revenue":revenue_list,
                       "Production Country":country_list,
                       "Release Date":release_date_list,
                       "Run Time":run_time_list,
                       "Vote Average":vote_ave_list,
                       "Vote Count":vote_count_list,
                       "imdb_id": imdb_id_list})
tmdb_movie_df.head()
```
