## General Clean Up Data and Save to CSV file
##### Drop columns with the same data info and convert the 0 values in Budget and Revenue Section to Null, in order to finalize the final Data frame for analyzing

```python
# # Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import sys
# Import API key
# from api_keys import api_key
# from api_keys2 import api_key2
```

#### Merge OMDB & TMDB From Excel files

```python
# Recall the OMDB data and TMDB data
df_tmdb = pd.read_csv("output_data/tmd_api_movie_2019.csv")
df_omdb = pd.read_csv("output_data/omdb_api_movie_2019.csv")
```

```python
#combine the OMDB and TMDB data to generate final dataframe
df_tmdb["omdb_id"] = df_tmdb["imdb_id"]
df_imdb = pd.merge(df_tmdb, df_omdb, on="omdb_id", how="left")
df_imdb.head()
```

```python
df_imdb.to_csv("output_data/imdb_api_movie_2019.csv",index=False)
```

#### Clean Up Final Data Frame

```python
#Get the column names to drop repeat columns
df_imdb.columns
```

```python
df_movie=df_imdb.drop(columns=["ID",
                               "Release Date",
                               "Language",
                               "imdb_id",
                               "omdb_id",
                               "omdb_title",
                               "omdb_year",
                               "omdb_ratings"])
df_movie.head()
```

```python
df_movie=df_movie.rename(columns={"omdb_rated":"Rated",
                         "omdb_director":"Director",
                         "omdb_actors":"Actors",
                         "omdb_language":"Language",
                         "omdb_metascore":"Metascore",
                         "omdb_imdbRating":"IMDB Rating",
                         "omdb_imdbVotes":"IMDB Votes"})
df_movie.head(1)
```

```python
#Convert 0 value in Budget and Revenue to NaN since 0 means missing data here
df_movie["Budget"] = df_movie["Budget"].replace(0,np.nan)
df_movie["Revenue"] = df_movie["Revenue"].replace(0,np.nan)
df_movie.head()
```

#### Cleaning Result Summary

```python
df_movie.info()
```

```python
df_movie.to_csv("output_data/cleanup_movie_2019.csv",index=False)
```
