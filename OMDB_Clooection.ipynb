```python
import csv
imdb_id_list = []

with open('output_data/tmd_api_movie_2019.csv', encoding="utf8") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    omdb_line_counter = 0
    for row in csv_reader:
        if omdb_line_counter == 0:
            omdb_line_counter = 1
        else:
            imdb_id_list.append(row[14])
            omdb_line_counter +=1
```

```python
#initial counter
record_counter2=0
set_counter2=1

#define list
omdb_id_list=[]
omdb_title_list=[]
omdb_year_list=[]
omdb_rated_list=[]
omdb_released_list=[]
omdb_runtime_list=[]
omdb_director_list=[]
omdb_writer_list=[]
omdb_actors_list=[]
omdb_language_list=[]
omdb_country_list=[]
omdb_ratings_list=[]
omdb_metascore_list=[]
omdb_imdbRating_list=[]
omdb_imdbVotes_list=[]
omdb_boxoffice_list=[]
omdb_production_list=[]

base_url = "http://www.omdbapi.com/?"

print("Beginning Data Retrieval From OMDB\n")
print("-"*30)
#Scrap data from api  http://www.omdbapi.com/?i=tt3896198&apikey=8243242d
for om_id in imdb_id_list:
    try:
        url=base_url+"i="+str(om_id)+"&apikey="+api_key2
        response=requests.get(url).json()
        omdb_id_list.append(response["imdbID"])
        omdb_title_list.append(response["Title"])
        omdb_year_list.append(response["Year"])
        omdb_rated_list.append(response["Rated"])
        omdb_released_list.append(response["Released"])
        omdb_runtime_list.append(response["Runtime"])
        omdb_director_list.append(response["Director"])
        omdb_actors_list.append(response["Actors"])
        omdb_language_list.append(response["Language"])
        omdb_country_list.append(response["Country"])
        omdb_ratings_list.append(response["Ratings"])
        omdb_metascore_list.append(response["Metascore"])
        omdb_imdbRating_list.append(response["imdbRating"])
        omdb_imdbVotes_list.append(response["imdbVotes"])
        omdb_boxoffice_list.append(response["BoxOffice"])
        omdb_production_list.append(response["Production"])
        
        record_counter2+=1
        sys.stdout.write('\r'+'Processing Record '+str(record_counter2))
        sys.stdout.flush()
    
    except:
        pass

print ("\nData Retrieval Complete\n")
print ("-"*30)
print ("Movie Found "+str(record_counter2))
```

```python
omdb_movie_df=pd.DataFrame({"omdb_id":omdb_id_list,
                       "omdb_title":omdb_title_list,
                       "omdb_year":omdb_year_list,
                       "omdb_rated":omdb_rated_list,
                       "omdb_released":omdb_released_list,
                       "omdb_runtime":omdb_runtime_list,
                       "omdb_director":omdb_director_list,
                       "omdb_actors":omdb_actors_list,
                       "omdb_language":omdb_language_list,
                       "omdb_country":omdb_country_list,
                       "omdb_ratings":omdb_ratings_list,
                       "omdb_metascore":omdb_metascore_list,
                       "omdb_imdbRating":omdb_imdbRating_list,
                       "omdb_imdbVotes":omdb_imdbVotes_list})

omdb_movie_df.head()
```
