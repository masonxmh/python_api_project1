```python
# # Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import sys
# Import API key
from api_keys import api_key
#Read csv data set
tmdb_5000_movies = "Resources/tmdb_5000_movies.csv"
release_dates_per_country="Resources/release_dates_per_country.csv"
movie_metadata="Resources/movie_metadata.csv"
tmdb_data=pd.read_csv(tmdb_5000_movies)
release_data=pd.read_csv(release_dates_per_country)
imdb_data=pd.read_csv(movie_metadata)
#output data path
output_data_file = "output_data/tmd_api_movie.csv"
```

```python
#Generate movie id list
movie_5000_id = tmdb_data["id"].tolist()
movie_release_id =release_data["movie_id"].tolist()
#Combine movie id list by unique values
movie_id=np.unique(movie_5000_id+movie_release_id)
#check how many ids in the movie_id list
print(len(movie_id))
```

```python
#initial counter
record_counter=0
set_counter=1

#define list
id_list=[]
title_list=[]
adult_list=[]
budget_list=[]
genres_list=[]
imdb_id_list=[]
language_list=[]
pop_list=[]
company_list=[]
country_list=[]
release_date_list=[]
revenue_list=[]
run_time_list=[]
vote_ave_list=[]
vote_count_list=[]

base_url = "https://api.themoviedb.org/3/movie/"

print("Beginning Data Retrieval\n")
print("-"*30)
#Scrap data from api
for m_id in movie_id:
    try:
        url=base_url+str(m_id)+"?api_key="+api_key#+"&language=en-US"
        response=requests.get(url).json()
        id_list.append(response["id"])
        title_list.append(response["title"])
        adult_list.append(response["adult"])
        budget_list.append(response["budget"])
        genres_list.append([i["name"] for i in response["genres"]])
        imdb_id_list.append(response["imdb_id"])
        language_list.append(response["original_language"])
        pop_list.append(response["popularity"])
        company_list.append([j["name"] for j in response["production_companies"]])
        country_list.append([k["name"] for k in response["production_countries"]])
        release_date_list.append(response["release_date"])
        revenue_list.append(response["revenue"])
        run_time_list.append(response["runtime"])
        vote_ave_list.append(response["vote_average"])
        vote_count_list.append(response["vote_count"])
        
        record_counter+=1
        sys.stdout.write('\r'+'Processing Record '+str(record_counter))
        sys.stdout.flush()
#         if record_counter==1000:  # test loop use only
#             break
#        if record_counter==51:
#             record_counter=1
#             set_counter+=1
#        print("Processing Record "+str(record_counter)+" | Movie ID"+ str(m_id))

    except:
        pass

#        print("Movie not found. Skipping ...")
#print ("-"*30)
print ("\nData Retrieval Complete\n")
print ("-"*30)
print ("Movie Found "+str(record_counter))
```

```python
#Generate tmdb data frame
tmdb_df=pd.DataFrame({ "imdb ID": imdb_id_list,
                         "ID":id_list,
                       "Title": title_list,
                       "Adult":adult_list,
                       "Budget":budget_list,
                           "Genres":genres_list,
                         "Language":language_list,
                          "Production Company":company_list,
                         "Popularity":pop_list,
                          "Revenue":revenue_list,
                            "Production Country":country_list,
                          "Release Date":release_date_list,
                          "Run Time":run_time_list,
                          "Vote Average":vote_ave_list,
                        "Vote Count":vote_count_list})
tmdb_df.head()
```

```python
#save data frame to csv
tmdb_df.to_csv("output_data/tmd_api_movie_original.csv",index=False)
```

```python
#clean up imdb data set pull out of the imdb id in order to merge with tmdb data set
imdb_df=imdb_data.drop(["color","duration","movie_title","facenumber_in_poster","language","country","budget","genres","num_critic_for_reviews","gross"],axis=1)
columnTitles=["movie_imdb_link","director_name","director_facebook_likes","actor_1_name","actor_1_facebook_likes","actor_2_name","actor_2_facebook_likes",
              "actor_3_name","actor_3_facebook_likes","plot_keywords","num_voted_users","cast_total_facebook_likes","num_user_for_reviews","content_rating","title_year",
              "imdb_score","aspect_ratio","movie_facebook_likes"]
imdb_df=imdb_df.reindex(columns=columnTitles)
imdb_df.head()
start='http://www.imdb.com/title/'
end='/?ref_=fn_tt_tt_1'
split1=[ elem.split(start)[1] for elem in imdb_df["movie_imdb_link"] ]
split2=[ elem.split(end)[0] for elem in split1]
imdb_df["imdb ID"]=split2
imdb_df=imdb_df.drop(["movie_imdb_link"],axis=1)
columnTitles_new=imdb_df.columns.tolist()
columnTitles_new=columnTitles_new[-1:]+columnTitles_new[:-1]
imdb_df=imdb_df[columnTitles_new]
imdb_df.head()
```

```python
#save data frame to csv
imdb_df.to_csv("output_data/imdb_movie_original.csv",index=False)
```

```python
#combine the tmdb and imdb data frame
movie_df=pd.merge(tmdb_df,imdb_df,on="imdb ID",how='inner')
movie_df.count()
movie_df.head()
movie_df.to_csv("output_data/combine_movie_original.csv",index=False)
```
